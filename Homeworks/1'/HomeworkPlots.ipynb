```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_layer1 = pd.read_excel("Homework1_results_II.xlsx")
df_layer1 = df_layer1.drop(0)
```

```python
df_layer1.head()
```

```python
test_accuracy = (np.array(df_layer1["Test Accuracy"]) * 100)
df_layer1["Test_Accuracy"] = test_accuracy
df_layer1.head()
```

```python
sns.set(style="whitegrid", font_scale = 0.6)
palette = sns.color_palette("mako_r", 3)
# plot1 = sns.lineplot(data=df_layer1, x="Size of Layer 1", y="Test_Accuracy",
#                      marker='o', hue="SIze of Layer 2", palette=palette)
plot1 = sns.lineplot(data=df_layer1, x="Total Number of Layers", y="Test_Accuracy",
                     marker='o', hue="Standard Size of the Other Fully Connected Layers", palette=palette)
plot1.set_xticks(df_layer1["Total Number of Layers"])
plot1.set_yticks(df_layer1["Test_Accuracy"])
plt.setp(plot1.get_legend().get_texts(), fontsize='5')

# for legend title
plt.setp(plot1.get_legend().get_title(), fontsize='5')
plt.savefig("homework1_II.png", dpi=300)
plt.show()
```
